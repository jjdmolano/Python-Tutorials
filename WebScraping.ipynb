```python
def test(**kwargs):
    for key, value in kwargs.items():
        if key == 'One':
            print('HIT')
            continue
        if key == 'Three':
            print('Buzz')
            continue
        else:
            print(value)
```

```python
my_dict = {'One':'1', 'Two':'2', 'Three':'3', 'Four':'4'}
```

```python
test(**my_dict)
```
